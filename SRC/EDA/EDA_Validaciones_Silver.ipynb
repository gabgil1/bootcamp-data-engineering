```sql

SELECT 
    (SELECT count(*) FROM bootcamp_gabriel.bronze.properties_bronze) as total_bronze,
    (SELECT count(*) FROM bootcamp_gabriel.silver.properties_silver) as total_silver,
    -- Calculamos la diferencia exacta
    (SELECT count(*) FROM bootcamp_gabriel.bronze.properties_bronze) - 
    (SELECT count(*) FROM bootcamp_gabriel.silver.properties_silver) as registros_eliminados_por_limpieza;
```

```sql
-- EJERCICIO 1: Ranking de Precios
-- Objetivo: Obtener las 5 propiedades más caras de la región 'gba zona norte'.
-- Requisito: Usar una CTE para filtrar la región y luego ordenar el resultado final.

with region_norte as( 
  select * from bootcamp_gabriel.silver.properties_silver where region = 'gba zona norte'
)
select * from region_norte order by precio desc limit 5;
```

```sql
with diferencia_metros as (
  select *,sum(metros_totales-metros_cubiertos) as diferencia_metroscol from bootcamp_gabriel.silver.properties_silver 
)
select * from diferencia_metros where diferencia_metroscol > 50 order by diferencia_metroscol desc;
```
