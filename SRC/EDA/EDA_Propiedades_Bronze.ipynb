```sql

DESCRIBE TABLE bootcamp_gabriel.bronze.properties_bronze;

SELECT 
    precio, expensas, ambientes, 
    metros_cuadrados_totales, metros_cuadrados_cubiertos, 
    antiguedad 
FROM bootcamp_gabriel.bronze.properties_bronze 
SUMMARY;
```

```sql
SELECT 
    count(*) as total_registros,
    sum(case when precio is null or precio = '0' then 1 else 0 end) as nulos_precio,
    sum(case when zona is null or zona = '' then 1 else 0 end) as nulos_zona,
    sum(case when moneda is null then 1 else 0 end) as nulos_moneda,
    sum(case when ambientes is null or ambientes = '0' then 1 else 0 end) as nulos_ambientes
FROM bootcamp_gabriel.bronze.properties_bronze;
```

```sql
-- ¿Qué monedas tenemos y cuántas de cada una?
SELECT moneda, count(*) as cantidad 
FROM bootcamp_gabriel.bronze.properties_bronze 
GROUP BY moneda;
```

```sql
SELECT tipo_vendedor, count(*) as cantidad 
FROM bootcamp_gabriel.bronze.properties_bronze 
GROUP BY tipo_vendedor;
```

```sql
SELECT zona, count(*) as cantidad 
FROM bootcamp_gabriel.bronze.properties_bronze 
GROUP BY zona 
ORDER BY cantidad DESC 
LIMIT 20;
```

```sql
SELECT 
    max(cast(ambientes as int)) as max_ambientes,
    min(cast(precio as decimal)) as min_precio,
    max(cast(metros_cuadrados_totales as decimal)) as max_metros
FROM bootcamp_gabriel.bronze.properties_bronze;
```
