```sql
CREATE OR REPLACE TABLE bootcamp_gabriel.silver.dim_zona (
    id_zona STRING,         -- PK generada (ej: MD5 de partido + region)
    zona_original STRING,
    partido STRING,
    region STRING
) USING DELTA;

-- 2. Dimensión de Propiedad (El "Qué" - Atributos fijos)
CREATE OR REPLACE TABLE bootcamp_gabriel.silver.dim_propiedad (
    id_propiedad STRING,    -- PK (ID original)
    ambientes INT,
    cochera STRING,
    antiguedad INT,
    url STRING
) USING DELTA;

-- 3. Dimensión de Tiempo (El "Cuándo")
CREATE OR REPLACE TABLE bootcamp_gabriel.silver.dim_tiempo (
    id_fecha DATE,          -- PK
    dia INT,
    mes INT,
    anio INT,
    trimestre INT,
    dia_semana STRING
) USING DELTA;

-- 4. Tabla de Hechos (El "Cuánto" - Métricas y Conexiones)
CREATE OR REPLACE TABLE bootcamp_gabriel.silver.fact_ventas (
    id_fact BIGINT,         -- PK de la transacción
    id_zona STRING,         -- FK a dim_zona
    id_propiedad STRING,    -- FK a dim_propiedad
    id_fecha DATE,          -- FK a dim_tiempo
    precio DECIMAL(15,2),
    metros_totales DECIMAL(10,2),
    metros_cubiertos DECIMAL(10,2),
    precio_m2 DECIMAL(15,2)
) USING DELTA;
```

```sql
INSERT OVERWRITE TABLE bootcamp_gabriel.gold.dim_tiempo
SELECT DISTINCT 
    CAST(fecha_proceso AS DATE) as id_fecha,
    day(fecha_proceso) as dia,
    month(fecha_proceso) as mes,
    year(fecha_proceso) as anio,
    quarter(fecha_proceso) as trimestre,
    date_format(fecha_proceso, 'EEEE') as dia_semana
FROM bootcamp_gabriel.silver.properties_silver;
```
